```python
import matplotlib.pyplot as plt

# Given values
results = [6, 4, 6, 4, 6, 6]
conditions = [f'Condition {i+1}' for i in range(len(results))]

# Create the plot
plt.figure(figsize=(8, 5))
plt.bar(conditions, results, color='skyblue')

# Set labels and title
plt.ylabel('Value')
plt.ylim(0, 9)  # Setting y-axis limit
plt.title('Correct per Condition')

# Display the plot
plt.show()
```
